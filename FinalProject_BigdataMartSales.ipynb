```python
# importing the required libraries for regression
from sklearn.preprocessing import *
from sklearn.linear_model import LinearRegression
from sklearn.tree import *
from sklearn.model_selection import train_test_split
from sklearn import model_selection
from sklearn.ensemble import *
from sklearn.metrics import *
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
X_train.drop(['Item_Identifier','Outlet_Identifier','Outlet_Establishment_Year'],axis=1, inplace=True)
X_test.drop(['Item_Identifier','Outlet_Identifier','Outlet_Establishment_Year'],axis=1, inplace=True)
```

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
```

```python
def score(model, X_train=X_train, Y_train=Y_train):
    print("Average R2 Score :", np.average(cross_val_score(model, X_train, Y_train, cv=10)))
    print("Average Root Mean Square Error :", np.average(cross_val_score(model, X_train, Y_train, cv=10, scoring='neg_root_mean_squared_error')))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

LR = LinearRegression(positive=True)
LR.fit(X_train,Y_train)
score(LR)
```

```python
score(LR,X_test, Y_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

poly = PolynomialFeatures(3)
X_poly = poly.fit_transform(X_train)

PR = LinearRegression(positive=True)
PR.fit(X_poly, Y_train)

score(PR, X_poly, Y_train)
```

```python
score(PR,X_test, Y_test)
```

```python
# polynomial regression --> second degree
from sklearn.preprocessing import PolynomialFeatures
poly_2_features = PolynomialFeatures(degree=2, include_bias=False)
mart_prepared_poly2 = poly_2_features.fit_transform(X_train)

poly_reg_2 = LinearRegression()
poly_reg_2.fit(mart_prepared_poly2, Y_train)
```

```python
housing_test_prepared_poly_2 = poly_2_features.transform(X_test)
housing_test_predictions_2 = poly_reg_2.predict(housing_test_prepared_poly_2)
poly_2_mse = mean_squared_error(Y_test, housing_test_predictions_2)
poly_2_rmse = np.sqrt(poly_2_mse)
poly_2_rmse
```
